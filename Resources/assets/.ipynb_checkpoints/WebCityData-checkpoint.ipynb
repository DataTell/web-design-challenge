```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import pandas as pd
import numpy as np
from scipy.stats import linregress
%matplotlib inline
```

```python
df = pd.read_csv("cities.csv")
df
```

```python
# Build a scatter plot for all cities vs max. temp.:
plt.figure(figsize=(10,7))
plt.scatter(df['Lat'], df['Max Temp (F)'], marker='o', edgecolors='black',s=[90],c=df["Max Temp (F)"],alpha=1,cmap='winter')#color='royalblue'

# Incorporate the other graph properties:
plt.title('City Latitude vs. Max Temperature (2020-04-11)',fontsize=20,pad=25) # ' + '(' + str(pd.datetime.now().date()) + ')' -for an automatic date in the title.
plt.ylabel('Max Temperature (F$^o$)', fontsize=16, color='black',labelpad=15)
plt.xlabel('Latitude', fontsize=16, color='black',labelpad=15)

# Set the limits of the y axis:
plt.ylim(min(df['Max Temp (F)'])-10, max(df['Max Temp (F)'])+10)

# Set the limits of the x axis:
plt.xlim(min(df['Lat'])-5, max(df['Lat'])+5)

# Save the figure:
plt.savefig('City Latitude vs. MaxTemp.png',dpi = 100)

# Show plot:
plt.show()
```

```python
# Build a scatter plot for all cities vs humidity level:
plt.figure(figsize=(10,7))
plt.scatter(df['Lat'], df['Humidity (%)'], marker='o', edgecolors='black',s=[90],c=df["Humidity (%)"],alpha=1,cmap='winter')

# Incorporate the other graph properties:
plt.title('City Latitude vs. Humidity (2020-04-11)', fontsize=20,pad=25)
plt.ylabel('Humidity (%)', fontsize=16, color='black',labelpad=15)
plt.xlabel('Latitude', fontsize=16, color='black',labelpad=15)

# Set the limits of the y axis:
plt.ylim(min(df['Humidity (%)'])-10, max(df['Humidity (%)'])+10)

# Set the limits of the x axis:
plt.xlim(min(df['Lat'])-5, max(df['Lat'])+5)

# Save the figure:
plt.savefig('City Latitude vs. Humidity.png',dpi = 100)

# Show plot:
plt.show()
```

```python
# Build a scatter plot for all cities vs cloudiness level:
plt.figure(figsize=(10,7))
plt.scatter(df['Lat'], df['Cloudiness (%)'],marker='o',edgecolors='black',s=[90],c=df["Cloudiness (%)"],alpha=1,cmap='winter')#Use cmap='winter_r' to rotate colors upsidedown.

# Incorporate the other graph properties:
plt.title('City Latitude vs. Cloudiness (2020-04-11)', fontsize=20,pad=25)
plt.ylabel('Cloudiness (%)', fontsize=16, color='black',labelpad=15)
plt.xlabel('Latitude', fontsize=16, color='black',labelpad=15)

# Set the limits of the y axis:
plt.ylim(min(df['Cloudiness (%)'])-10, max(df['Cloudiness (%)'])+10)

# Set the limits of the x axis:
plt.xlim(min(df['Lat'])-5, max(df['Lat'])+5)

# Save the figure
plt.savefig('City Latitude vs. Cloudiness.png',dpi = 100)

# Show plot
plt.show()
```

```python
# Build a scatter plot for all cities vs wind speed:
plt.figure(figsize=(10,7))
plt.scatter(df['Lat'], df['Wind Speed (mph)'], marker='o', edgecolors='black',s=[90],c=df["Wind Speed (mph)"],alpha=1,cmap='winter')

# Incorporate the other graph properties:
plt.title('City Latitude vs. Wind Speed (2020-04-11)', fontsize=20,pad=25)
plt.ylabel('Wind Speed (mph)', fontsize=16, color='black',labelpad=15)
plt.xlabel('Latitude', fontsize=16, color='black',labelpad=15)

# Set the limits of the y axis:
plt.ylim(min(df['Wind Speed (mph)'])-1, max(df['Wind Speed (mph)'])+5)

# Set the limits of the x axis:
plt.xlim(min(df['Lat'])-5, max(df['Lat'])+5)

# Save the figure
plt.savefig('City Latitude vs. Wind.png',dpi = 100)

# Show plot
plt.show()
```
